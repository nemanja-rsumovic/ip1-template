```python
import pandas as pd
import numpy as np
from matplotlib.image import imread
from matplotlib import pyplot as plt
```

```python
img = imread('dog.jpg')
img_size = img.shape
img_size
```

```python
img = img.reshape(img_size[0] * img_size[1], img_size[2])
img
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=20, n_init='auto')
```

```python
kmeans.fit(img)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_ #niz labela
```

```python
X_compressed = kmeans.cluster_centers_[kmeans.labels_]
X_compressed = np.clip(X_compressed.astype('uint8'), 0, 255)
X_compressed = X_compressed.reshape(img_size[0],img_size[1],img_size[2])
X_compressed
```

```python
plt.imshow(X_compressed)
```
