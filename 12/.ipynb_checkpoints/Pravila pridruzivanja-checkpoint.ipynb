```python
import pandas as pd
```

```python
# učitavanje skupa
dataset = pd.read_excel('IP1-materijali-private/2021_2022/12/izborni_predmeti.xlsx')
dataset
```

```python
# Filtriramo studente I i R smera
f = lambda x: x['NAZIV_SMERA'] in ['Informatika', 'Racunarstvo i informatika']  
dataset = dataset[dataset.apply(f, axis=1)]
```

```python
# Grupisanje po indeksu
import numpy as np
dataset = dataset.groupby(['INDEKS'])['NAZIV_PREDMETA'].apply(list).values.tolist()
dataset
```

```python
# Apriori za min sup 2% i min conf 80%
from apriori_python import apriori
freqItem, rules = apriori(dataset, minSup=0.02, minConf=0.80)
```

```python
pd.DataFrame(rules, columns=['Antecedent', 'Consequent', 'Confidence'])
```
