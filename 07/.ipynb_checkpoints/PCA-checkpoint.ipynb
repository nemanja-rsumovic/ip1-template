# PCA

Analiza glavnih komponenti (_Principal Component Analysis_, PCA) je nenadgledana procedura za izgradnju novih linearno nekorelisanih atributa od postojecih. Te nove atribute zovemo glavne komponente. Prva glavna komponenta je vektor koji ima najvecu mogucu varijansu, tj. duz ovog pravca podaci najvise variraju. Svaka sledeca glavna komponenta je normalna u odnosu na sve prethodne i opet objasnjava najveci moguci preostali deo varijanse podataka.

Nakon ove transformacije mozemo da uzmemo prvih k glavnih komponenti i na taj nacin smanjimo dimenzionalnost skupa podataka, a da izgubimo sto manji deo informacija.

(_procedura za smanjenje dimenzionalnosti podataka_)

Odabir glavnih komponenti koje imaju maksimalnu varijansu omogućava nam da identifikujemo pravce u prostoru podataka duž kojih su podaci najviše raspršeni, najviše variraju. To znači da te glavne komponente zadržavaju što više informacija o raspodeli podataka i njihovoj strukturi. Kada smanjujemo dimenzionalnost podataka, želimo zadržati što više bitnih informacija. Odabir glavnih komponenti s maksimalnom varijansom osigurava da zadržimo što veći deo varijabilnosti podataka, što rezultira efikasnijim smanjenjem dimenzionalnosti uz očuvanje bitnih karakteristika podataka.

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
data = load_breast_cancer()
```

```python
X = data['data']
y = data['target']
feature_names = data['feature_names']
class_names = data['target_names']
```

```python
X.shape
```

```python
y.shape
```

Sam PCA je nenadgledan, ali ako zelimo da primenimo neki klasifikacioni algoritam nakon njega, moramo prvo da podelimo na trening i test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

Preduslov za primenu PCA je da svi atributi imaju prosek 0 (**&mu;=0** ali u redu je i da bude **&sigma;=1** da bi koristili _StandardScaler_)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=feature_names)
X_test = pd.DataFrame(X_test, columns=feature_names)
```

```python
from sklearn.decomposition import PCA
```

Za argument _n_components_ možemo da pošaljemo:
- _int_ - koliko glavnih komponenti želimo
- _float_ - koliki udeo varijanse želimo da objasnimo

--- 
_n_components=2_
  
* pca_0 -> 0.6

* pca_1 -> 0.25  [dve glavne komponente i tu staje]

--- 
_n_components=0.95_
  
* pca_0 -> 0.6

* pca_1 -> 0.25

* pca_3 -> 0.11  [pređen udeo od 0.95 i tu staje]

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X_train)
```

```python
pca.components_.shape
```

Glavne komponente su linearne kombinacije polaznih atributa i clanska promenljiva _components__ nam daje koeficijente

```python
pca_names = [f'Principal_component_{i}' for i in range(pca.n_components_)]
pca_names
```

```python
for i, component in enumerate(pca.components_):
    print(f'{pca_names[i]} = ', end='')
    print(' + '.join(f'{coef:.2f}*{feature_name}'
                     for coef, feature_name in zip(component, feature_names)))
    print('-'*80)
```

Ili vizuelno

```python
from matplotlib import pyplot as plt
```

```python
plt.matshow(pca.components_)
plt.xticks(ticks=range(pca.components_.shape[1]),
           labels=feature_names,
           rotation='vertical')
plt.yticks(ticks=range(pca.n_components_),
           labels=pca_names)
plt.colorbar()
plt.show()
```

Objasnjena varijansa svake od glavnih komponenti

```python
pca.explained_variance_
```

Udeo objasnjene varijanse svake od glavnih komponenti

```python
pca.explained_variance_ratio_
```

Ili vizuelno

```python
plt.bar(x=pca_names,
        height=pca.explained_variance_ratio_)
plt.title('Udeo objasnjene varijanse svake od glavnih komponenti pojedinacno')
plt.show()
```

```python
import numpy as np
```

Kumulativni udeo varijanse:

_cumulative_variance_ratio = [pca_0, pca_0 + pca_1]_

```python
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
cumulative_variance_ratio
```

```python
plt.plot(pca_names, cumulative_variance_ratio, marker='o')
plt.title('Kumulativna suma udela objasnjene varijanse')
plt.show()
```

Prve dve glavne komponente objasnjavaju ukupno 63% ukupne varijanse podataka

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
X_train_pca.shape
```

Dve glavne komponente su nasi novi atributi

```python
from matplotlib import pyplot as plt
```

```python
scatter = plt.scatter(X_train_pca[:,0], X_train_pca[:,1], c=y_train)
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.legend(handles=scatter.legend_elements()[0], labels=class_names.tolist())
plt.show()
```

Mozemo da odredimo vrednosti polaznih atributa na osnovu vrednosti novih atributa

```python
pca.inverse_transform(np.random.random(size=(1,2)))
```

## Linearni (pot)prostor

Kako je svaka glavna komponenta linearna kombinacija polaznih atributa i međusobno su ortogonalne, na ovaj način redukujemo polazni visokodimenzioni prostor na linearni potprostor. Dakle, _suštinska_ dimenzionalnost naših podataka može biti manja od one koju pronađemo korišćenjem PCA jer će nam on uvek dati _linearan potprostor_.

Uzmimo primer u kojem su podaci raspoređeni duž površine sfere. Iako je ovo dvodimenzionalan prostor, nije linearan u smislu da ne možemo jednostavno iscrtati pravu liniju ili ravan koja bi odvojila podatke na dva različita dela.

PCA, kao linearna tehnika, može imati poteškoće u identifikovanju nelinearnih struktura u podacima, kao što je površina sfere i u takvim slučajevima neće moći da pronađe odgovarajući linearni potprostor. Za identifikaciju ovakvih nelinearnih struktura u podacima potrebne su naprednije tehnike analize podataka, poput nelinearnih metoda redukcije dimenzionalnosti, kao što su _Kernel PCA_ ili metode bazirane na neuronskim mrežama.

```python
r = 1
n = 25
phis = np.linspace(0, 180, num=n)
thetas = np.linspace(0, 360, num=n)
x = r * np.outer(np.sin(phis), np.cos(thetas))
y = r * np.outer(np.sin(phis), np.sin(thetas))
z = r * np.outer(np.cos(phis), np.ones(n))
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z)
plt.show()
```

```python
data = np.concatenate((x.reshape(-1,1), y.reshape(-1,1), z.reshape(-1,1)), axis=1)
data.shape
```

```python
pca = PCA()
```

```python
pca.fit(data)
```

```python
pca.explained_variance_ratio_
```
